## Dataset 1: glacial data

Begin with importing the data:

# `plot()` method

`pandas.DataFrame` has a built in method `plot()` for plotting.
Default without specifying parameters:

- creates 1 line plot for each of the columns with numeric data
- sets the values of the index as the x-axis values 

Example:

````python
# default plot - a mess!!
df.plot()
````

## Line Plots

Make line plots of 1 column against another with the following syntax:

```
df.plot(x = 'x_values_column', y = 'y_values_column')
```

Example:

````python
# change in galcial volume per year in Europe - simple plot!

df.plot(x = 'year',
       y = 'europe')
````

Basic customization specifying other arguments in `plot()`

Example:


````python
# change in galcial volume per year in Europe - simple plot!

df.plot(x = 'year',
        y = 'europe',
       title = 'Change in Glacial Volume per Year in Europe',
       xlabel = 'Year',
       ylabel = 'Change in Glacial Volume (km3)',
       color = '#FFB900FF')
````

# Multiple line Plots

Want to graph the change in glacial volumn in the Arctic and Alaska. Update these arguments:

- 'y': a list of column names that will be plotted against x
- 'color': specify the color of each column with a dictionary `{'col1':'color1', 'col2':'color2'`}

Example:

````python
# color dictionary
colors = {
    'arctic' : '#5C88DAFF',
    'alaska' : '#FFA400FF'
}

df.plot(x = 'year',
        y = ['arctic', 'alaska'],
        color = colors)
````

Create separate plots setting the `subplots` feature argument to `True`.

````python
df.plot(x = 'year',
        y = ['arctic', 'alaska'],
        color = colors,
        # separare the plots into different frames
       subplots = True)
````
